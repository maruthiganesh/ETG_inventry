```python
#item_id,itemname,category,cost,stock_available,discount

import json

total=0
user_cart={}

fd=open("Inventory.txt","r")
txt=fd.read()
d=json.loads(txt)

print("---------------------------------------------------------------------------------")
print ("{:<11} {:<26} {:<12} {:<12} {:<14} {:<16}".format('item id','item','Category','Cost','Stock available','Discount in %'))
for k in sorted(d.keys()) :
    print("{:<11} {:<26} {:<12} {:<12} {:<14} {:<16}".format(k,d[k]["item_name"],d[k]["category"],d[k]["cost"],d[k]["stock_available"],d[k]["discount"]))

print("---------------------------------------------------------------------------------\n")

print("      Please enter item ids in one line separating with single space")
l=set(map(str, input().split()))
k=d.keys()
do=d.copy()
flag=0
for i in l:

    if(i in k):
        print("Enter quantity for",do[i]["item_name"])
        q=int(input())
        if(q<= do[i]["stock_available"]):
            user_cart[i]=do[i].copy()
            user_cart[i].pop("stock_available")
            user_cart[i]["Quantity"]=q
            do[i]["stock_available"]=do[i]["stock_available"]-q
            subt=user_cart[i]["cost"]*q
            disco=user_cart[i]["discount"]
            subt=subt*((100-disco)/100)
            user_cart[i]["price"]=subt
            total+=subt
        else:
            flag=1
            print("--------------------ITEM ",do[i]["item_name"]," is unavailable-----------------------")
if(flag==0):
    print("-----------------------------------------------Invoice----------------------------------------------------")
    print ("{:<26} {:<10} {:<10} {:<10} {:<10}".format('item','Cost','Quantity','Discount','Price'))
    for k in sorted(user_cart.keys()) :
        print ("{:<26} {:<10} {:<10} {:<10} {:<10}".format(user_cart[k]["item_name"],user_cart[k]["cost"],user_cart[k]["Quantity"],user_cart[k]["discount"],user_cart[k]["price"]))
    print("\nTotal Price:",total)
fd.close()
fp=open("Inventory.txt","w")


d=json.dumps(do)
fp.write(d)
fp.close()
```
